```python
import cv2
import time
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
cap = cv2.VideoCapture(0)
cv2.namedWindow('Original', cv2.WINDOW_NORMAL)
```

```python
ret, frame = cap.read()
img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
blur = cv2.GaussianBlur(img,(5,5),3)
gray = cv2.cvtColor(blur, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)
th3 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)
# ret, res = cv2.threshold(th3, 70, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
res = cv2.cvtColor(th3, cv2.COLOR_GRAY2RGB)

plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.imshow(img)
plt.subplot(1,3,2)
plt.imshow(th3)
plt.subplot(1,3,3)
plt.imshow(res)
plt.show()
```

```python
cap.release()
```
